```python
import pyautogui as gui
import time
import pyperclip as clip
import pandas as pd

link = 'https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga'

gui.PAUSE = 2 # delay no gui

gui.hotkey('ctrl', 't') # abre uma nova guia

clip.copy(link)
gui.hotkey('ctrl', 'v') 
gui.press('enter') # cola e abre o link

gui.click(431, 290, clicks=2) # clica na pasta
gui.click(430, 290) # clica no arquivo
gui.click(1707, 157) # clica nos três pontos
gui.click(1513, 555) # faz o download

time.sleep(10) # tempo para realizar o download

tabela = pd.read_excel(r"C:\Users\Raul\Downloads\Vendas - Dez.xlsx")
faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum() # lê a tabela baixada e faz a análise dos dados

link = "https://mail.google.com/mail/"

gui.hotkey('ctrl', 't')
clip.copy(link)
gui.hotkey('ctrl', 'v') 
gui.press('enter') # abre o gmail

gui.click(34, 174) # clica em escrever

gui.write("raul")
gui.press("tab") # escolher remetente

gui.press("tab") # passar para o campo de assunto
gui.write("Vendas") # digitar assunto

gui.press("tab") # passar para o campo de escrever o email

texto_email = f"""

Prezados,

O faturamento foi de: R$ {faturamento:,.2f}
A quantidade de produtos foi de: R$ {quantidade:,}

Att,
Raul
"""

clip.copy(texto_email)
gui.hotkey("ctrl", "v") # cola o corpo do email

gui.hotkey("ctrl", "enter") # envia o email


# gui.position() mostra a posição do mouse
```
